```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import matplotlib.image as mpimg

# Đường dẫn đến các tập tin ảnh
image_paths = ['/kaggle/working/Rouge Score Word2Vec L.png', '/kaggle/working/Rouge Score GloVe L.png', '/kaggle/working/Rouge Score FastText L.png', '/kaggle/working/Rouge Score TF-IDF L.png']

# Khởi tạo lưới 2x2
grid = GridSpec(2, 2)

# Chèn ảnh vào từng ô trong lưới
fig = plt.figure(figsize=(12, 8))  # Đặt kích thước của hình ảnh

for i, path in enumerate(image_paths):
    ax = fig.add_subplot(grid[i // 2, i % 2])  # Chỉ số của ô trong lưới
    img = mpimg.imread(path)
    ax.imshow(img)
    ax.axis('off')

plt.tight_layout()  # Tạo khoảng trắng tự động giữa các ô
plt.savefig('Rouge Scor 4 L.png')
plt.show()
```
